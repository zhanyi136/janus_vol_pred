```python
import pandas as pd
import zstandard as zstd

file_path = "/data/market_data/binance/future/XRPUSDT/book_ticker/binance-futures_book_ticker_2026-03-31_XRPUSDT.csv.zst"

with open(file_path, 'rb') as f:
    dctx = zstd.ZstdDecompressor()
    stream_reader = dctx.stream_reader(f)
    df = pd.read_csv(stream_reader)
    stream_reader.close()

df
```
